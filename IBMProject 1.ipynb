```python
del matches["comp"]
```

```python
del matches["notes"]
```

```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches["date"]
```

```python
#XG BOOST
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Select predictors and target variable
predictors = ["venue_code", "opp_code", "hour", "day_code", "gf", "ga"]
target = "target"  # Assuming "outcome" is the target variable

test = matches[matches["date"] > '2021-01-01']
train = matches[matches["date"] < '2021-01-01']

# Prepare train and test data
X_train = train[predictors]
y_train = train[target]
X_test = test[predictors]
y_test = test[target]

# Convert data to DMatrix format
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Define XGBoost parameters
params = {
    'objective': 'binary:logistic',  # Binary classification
    'eta': 0.1,  # Learning rate
    'max_depth': 3,  # Depth of each tree
    'eval_metric': 'logloss'  # Evaluation metric
}

# Train the XGBoost model
num_round = 10  # Number of boosting rounds
bst = xgb.train(params, dtrain, num_round)

# Predict
preds_proba = bst.predict(dtrain)
preds_binary = [1 if x > 0.5 else 0 for x in preds_proba]

# Calculate accuracy
accuracy = accuracy_score(y_train, preds_binary)
print("Accuracy: %.2f%%" % (accuracy * 100))
```
